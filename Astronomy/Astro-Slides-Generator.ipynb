# Create Slides per Chapter
The goal of this notebook is to process table of contents from the openstax astronomy and generate reveal js presentations of each chapter. Each chapter will be its own html page and the subsections of the book will be revealjs sections

```python
import requests
from bs4 import BeautifulSoup
import os
```

## Read the website for links

```python
url="https://openstax.org/books/astronomy/pages/1-introduction"
page=requests.get(url)
sp=BeautifulSoup(page.text,'html.parser')
```

```python
# Make list of string of section numbers and corresponding website link to text
links=[]
for ii in sp.find_all('a'):
    if '.' in ii.getText():
        links.append([ii.getText(),'https://openstax.org/books/astronomy/pages/'+ii.get('href')])
```

```python
# Find chapter title
chap_titles=[]
for ii in sp.find_all(class_='styled__SummaryTitle-sc-18yti3s-0 cSTpUE'):
    if ii.getText().split(' ')[0].isdigit():
        chap_titles.append(ii.getText())
```

```python
# Make directories based on chapter title names
print('****MAKE CERTAIN YOU WANT TO WRITE OVER EVERY HTML PAGE****')
WRITE_OVER_FILE=False
for tt in chap_titles:
    if int(tt[0:2])<10:
        file_name=('%02d-'%int(tt[0:2])+tt[2:].replace(' ','_').replace(':','').replace(',',''))
    else:
        file_name=('%02d-'%int(tt[0:2])+tt[3:].replace(' ','_').replace(':','').replace(',',''))
    # create directory if it does not already exist
    if not os.path.isdir(file_name):
        os.mkdir(file_name)

        
    # find path to 1st image of each chapter
    #------------------------------------------------------------------------------------------------------------------------#
    if int(file_name.split('-')[0])==1:
        page_image_url='https://openstax.org/books/astronomy/pages/1-introduction'
    else:
        page_image_url='https://openstax.org/books/astronomy/pages/'+str(int(file_name.split('-')[0]))+'-thinking-ahead'
        
    sp=BeautifulSoup(requests.get(page_image_url).text,'html.parser')
    images_in_page=sp.find_all('img')
    image_url_src='https://openstax.org'+images_in_page[1].get('src')

    # read in html template file and start editing it
    #------------------------------------------------------------------------------------------------------------------------#
    with open('00-template/Astro-Page-Template.html','r') as html:
        lines=html.readlines()
    html_template=''.join(lines)
    html_template=html_template.replace('HTML_TITLE',file_name.replace('_',' ').replace('-',': '))
    html_template=html_template.replace('SLIDES_TITLE','Chapter '+file_name.replace('_',' ').replace('-',': ').replace('0',''))
    html_template=html_template.replace('LINK_TO_CHAPTER_IMAGE',page_image_url)
    html_template=html_template.replace('THE_CHAPTER_IMAGE_SOURCE',image_url_src)
    
    # scan links list and create subsection slides
    subsec_slide=''
    for ll in links:
        if str(int(file_name.split('-')[0])) == ll[0].split('.')[0]:
            subsec_slide=subsec_slide+'    <section>\n        <section class="center">\n        <h2 class="overl"><a href='+ll[1]+' target="_blank"> <font style="color: white;">'+ll[0]+'</font></a></h2></section>'
            subsec_slide=subsec_slide+('\n            <section> <h3 class="overl">Subsection slide</h3>\n        </section>\n        </section>\n    </section>')
    html_template=html_template.replace('CHAPTER_SUBSECTIONS_SLIDES',subsec_slide)
    
    print('- ['+tt+'](Astronomy/'+file_name+'/'+file_name+'.html)')
    if WRITE_OVER_FILE:
        with open(file_name+'/'+file_name+'.html','w',encoding="utf-8") as new_chap_webpage:
            new_chap_webpage.write(html_template)
```

```python
with open(file_name+'/'+file_name+'.html','w',encoding="utf-8") as new_chap_webpage:
    new_chap_webpage.write(html_template)
```
