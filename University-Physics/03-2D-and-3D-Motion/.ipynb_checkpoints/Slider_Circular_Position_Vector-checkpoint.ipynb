# Slider Circular Position Vector

```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
```

```python
# Define physical parameters
R=1
w=1
t=np.linspace(0,2*np.pi,100)
```

```python
fig=go.Figure()

for tt in t:
    fig.add_trace(go.Scatter(
                  visible=False,
                  name=str(tt),
                  x=[0,np.cos(tt)],y=[0,np.sin(tt)],
                  line=dict(color="#dc5050", width=3)
                            ))
    fig.add_trace(go.Scatter(x=np.cos(w*t),y=np.sin(w*t),
                  line=dict(color="#000000", width=2)
                 ))
    
    
fig.data[0].visible = True
    
# Create slider
steps=[]
for ii in range(len(fig.data)):
    step = dict(
        method="update",
        args=[{"visible": [False] * len(fig.data)},
              {"title": "Slider switched to step: " + str(i)}],  # layout attribute
    )
    step["args"][0]["visible"][ii] = True  # Toggle i'th trace to "visible"
    steps.append(step)
    
sliders = [dict(
    active=0,
    currentvalue={"prefix": "Time: "},
    pad={"t": 100},
    steps=steps
)]

fig.update_layout(
    sliders=sliders
)


fig.update_layout(showlegend=False)
fig.update_yaxes(
    scaleanchor="x",
    scaleratio=1,
)
fig.show()
```
