# Energy Conservation Plot
Create series of images to show a particle oscilating from potential energy to kinetic energy
$$E=\tfrac{1}{2}mv^2+\tfrac{1}{2}kx^2$$
$$m\dot{x}=-kx^2$$
Given ICs $x(t=0)=x_0$ and $\dot{x}(t=0)=0$ we have
$$x(t)=x_0 \cos(\omega t), \quad \omega=\sqrt{\frac{k}{m}}$$

```python
import  numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
tmax=2*np.pi
t=np.linspace(0,tmax,1000)
stepsize=int(np.floor(len(t)/48))
x=np.cos(t)
v=-np.sin(t)
KE=1/2*v**2
Usp=1/2*x**2
E=KE+Usp
```

```python
stepsize
```

```python
#for ii in [30]:
for ii in range(len(t)-1):
    plt.close()
    fig, ax = plt.subplots(1,2)
    ax[0].plot(x,Usp,linewidth=2)
    #ax[0].plot(x,E,linewidth=2,color='r')
    ax[0].plot(x[ii*stepsize],Usp[ii*stepsize],marker='o',color='k',markersize=12)
    ax[0].axis('off')
    ax[1].plot(t[0:ii*stepsize],Usp[0:ii*stepsize])
    ax[1].plot(t[ii*stepsize],Usp[ii*stepsize],marker='o',color='k',markersize=12)
    ax[1].plot(t[0:ii*stepsize],KE[0:ii*stepsize])
    ax[1].plot(t[0:ii*stepsize],E[0:ii*stepsize])
    ax[1].axis([0,max(t),0,max(E)*1.05])
    plt.savefig('energy-oscilate-plot/%03d.png'%ii,bbox_inches='tight')
```
