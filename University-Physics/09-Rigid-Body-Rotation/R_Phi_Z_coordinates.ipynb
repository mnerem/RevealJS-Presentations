```python
import numpy as np
import pandas as pd
import plotly.express as px
#radians=np.arange(0,2*np.pi+2*np.pi/100,2*np.pi/100)
Angle=np.arange(0,360+1)
f=pd.DataFrame.from_dict(
    {
        'Angle': Angle,
        'X': np.cos(Angle*np.pi/180),
        'Y': np.sin(Angle*np.pi/180),
        'Size': .25*np.ones(np.size(Angle))
    }
)
fig=px.scatter(f,x='X',y='Y',size='Size',animation_frame='Angle',range_x=[-1.1,1.1],range_y=[-1.1,1.1],width=600,height=600)
fig.add_scatter(x=f.X,y=f.Y,name='Arc',showlegend=False)
fig.write_html('animated_arc.html')
fig.show()
```
